**Importing the test_cricket.xlsx file. And reading the sheet 'wickets' from there.**

```python
import pandas as pd
import numpy as np

df = pd.read_excel("test_cricket.xlsx", sheet_name='wickets', engine="openpyxl")

print("The first five rows of the data frame are:")
display(df.head(5))

print("The last five rows of the data frame are:")
display(df.tail(5))
```

**Displaying the first 10 rows of the dataframe**

```python
# Printing first 10 rows.
display(df.head(10))
```

**Column**

```python
# Printing all columns name
print(df.columns)
```

**Meaning of the Each Column:** 

    1. Player: The names of the players, played test cricket.

    2. Span: The career length of the players.

    3. Mat: Number of matches played by the players.

    4. Inns: Total innings played by the players.
    
    5. Balls: Number of balls bowled by the players. 
    
    6. Runs: Total runs earned by the players.
    
    7. Wkts: Total wickets earned by the players.
    
    8. BBI: Best bowling by the players in the innings.
    
    9. BBM: = Best bowling in the matches played by the players. 
    
    10. Ave = Average runs cost by the players per wicket
    
    11. Econ = The average number of runs, the players have conceded per over bowled.
    
    12. SR = Strike rate of the players. 
    
    13. 5 = Number of times the players earned 5 wickets.
    
    14. 10 = Number of times the players earned 10 wickets.

**Checking for missing values and data types of each column in the dataset.**

```python
# Deatil overview of the Dataset
print(df.info())

#Checking, if there are any null value in the columns.
print('Checking if there are any null value in the columns:')
df.isnull()
```

```python
# Drop a row if have at least 1 null value in any of it's columns. 
new_df = df.dropna(axis = 0, how ='any')

# Display the new dataset, without any null value of any columns. 
display(new_df.head(5))
print("Old data set length:", len(df))
print("New data set length:", len(new_df))

# Count the rows has at least 1 null values in any of it's columns. 
print("Number of rows with at least 1 Null value: ", (len(df)-len(new_df)))

# Count non-null values of each columns 
df.count()
```

**Rename the column names appropriately**

```python
 # Again printing columns' names to renamimng some the columns' names
print("Columns of the Dataset: ", df.columns)
```

```python
# Renamimng the columns' names which are a little bit hard to understand
df = df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        'Wkts': 'Wickets',
                        'BBI': 'Best_Bowling/Innings',
                        'BBM': 'Best_Bowling/Match',
                        'Ave': 'Average',
                        'Econ': 'Economy_Rate',
                        'SR': 'Strike_Rate',
                        5: '5_Wickets',
                        10: '10_Wickets'})

display(df.head())
```

**Remove a column from the dataframe**

```python
# Remove 10_Wickets columns as it has some null values in the rows.
df.drop('10_Wickets', axis=1, inplace=True)

display(df.head())
```
